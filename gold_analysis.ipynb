Bu projede, 19 Haziran 2024 - 19 Haziran 2025 tarihleri arasÄ±ndaki **gerÃ§ek altÄ±n vadeli iÅŸlem verileri** kullanÄ±larak fiyat trendleri analiz edilmiÅŸtir. 

### YapÄ±lanlar:
- GÃ¼nlÃ¼k fiyat grafiÄŸi oluÅŸturuldu.
- 20 gÃ¼nlÃ¼k ve 50 gÃ¼nlÃ¼k **Hareketli Ortalama** Ã§izgileri eklendi.
- Fiyat deÄŸiÅŸimleri grafiksel olarak gÃ¶rselleÅŸtirildi.
- Grafik renkleri ve Ã§izgi kalÄ±nlÄ±klarÄ± Ã¶zel olarak ayarlandÄ±.

Bu analiz, fiyatlardaki kÄ±sa ve uzun vadeli eÄŸilimleri gÃ¶zlemlemek iÃ§in temel bir teknik analiz Ã¶rneÄŸidir.


```python
import pandas as pd
import matplotlib.pyplot as plt

# DosyayÄ± oku
df = pd.read_csv("Gold Futures Historical Data.csv")

# Tarih sÃ¼tununu datetime yap 
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')

# Tarihi indeks yap
df.set_index('Date', inplace=True)

# VirgÃ¼lleri kaldÄ±rÄ±p sayÄ±ya Ã§evir
df['Price'] = df['Price'].str.replace(',', '').astype(float)

# NaN satÄ±rlarÄ± Ã§Ä±kar 
df.dropna(subset=['Price'], inplace=True)

df['MA20'] = df['Price'].rolling(window=20).mean()
df['MA50'] = df['Price'].rolling(window=50).mean()

# Grafik Ã§izimi
plt.figure(figsize=(14,7))
plt.plot(df.index, df['Price'], label='GÃ¼nlÃ¼k Fiyat', color='#FF69B4', linewidth=3.5)     
plt.plot(df.index, df['MA20'], label='20 GÃ¼nlÃ¼k Hareketli Ortalama', color='#800080')    
plt.plot(df.index, df['MA50'], label='50 GÃ¼nlÃ¼k Hareketli Ortalama', color='#1E90FF') 
plt.title('AltÄ±n FiyatlarÄ± ve Hareketli Ortalamalar')
plt.xlabel('Tarih')
plt.ylabel('Fiyat (USD)')
plt.legend()
plt.grid(True)
plt.show()
```
